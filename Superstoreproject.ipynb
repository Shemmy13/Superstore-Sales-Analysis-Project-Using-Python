```python
#Which month records the highest sales?
import pandas as pd
```

```python
df["Order.Date"] = pd.to_datetime(df["Order.Date"])
```

```python
df["Order.Date"]
```

```python
df["Month"] = df["Order.Date"].dt.month_name()
```

```python
df["Month"]
```

```python
maxmonthly_sales = df.groupby("Month")["Sales"].sum().idxmax()
```

```python
maxmonthly_sales
```

```python
df.groupby("Month")["Sales"].sum().sort_values(ascending=False)
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind="line", figsize=(10,5), marker="o")
plt.title("Sales Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
